4.Write a program that checks if a given word is a Python keyword.

```python
import keyword

def is_python_keyword(word):
    """
    Checks if a given word is a Python keyword.

    Args:
        word: The word to check.

    Returns:
        True if the word is a keyword, False otherwise.
    """
    return keyword.iskeyword(word)

# Example usage
word1 = "for"
word2 = "example"

print(f"'{word1}' is a keyword: {is_python_keyword(word1)}")
print(f"'{word2}' is a keyword: {is_python_keyword(word2)}")
```

6.Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def modify_data(immutable_arg, mutable_arg):
    """
    Demonstrates the behavior of mutable and immutable arguments in Python.

    Args:
        immutable_arg: An immutable argument (e.g., integer, string, tuple).
        mutable_arg: A mutable argument (e.g., list, dictionary).
    """
    print("Inside the function:")

    # Attempt to modify the immutable argument (creates a new object)
    immutable_arg = immutable_arg + 1
    print(f"  Immutable argument (inside): {immutable_arg}")

    # Modify the mutable argument (modifies the original object)
    mutable_arg.append(4)
    print(f"  Mutable argument (inside): {mutable_arg}")

# Example usage:
immutable_data = 10
mutable_data = [1, 2, 3]

print("Before function call:")
print(f"  Immutable data: {immutable_data}")
print(f"  Mutable data: {mutable_data}")

modify_data(immutable_data, mutable_data)

print("After function call:")
print(f"  Immutable data: {immutable_data}")
print(f"  Mutable data: {mutable_data}")
```

9.Write a Python program to convert user input from string to integer, float, and boolean types.

```python
def convert_string_to_types(input_string):
    try:
        integer_value = int(input_string)
    except ValueError:
        integer_value = None

    try:
        float_value = float(input_string)
    except ValueError:
        float_value = None

    if input_string.lower() == "true":
        boolean_value = True
    elif input_string.lower() == "false":
        boolean_value = False
    else:
        boolean_value = None

    return integer_value, float_value, boolean_value

input_string = input("Enter a value: ")
integer_value, float_value, boolean_value = convert_string_to_types(input_string)

print("Integer value:", integer_value)
print("Float value:", float_value)
print("Boolean value:", boolean_value)
```

14.Write a program to reverse a string using a while loop.

```python
def reverse_string(s):
    reversed_s = ""
    length = len(s) - 1
    while length >= 0:
        reversed_s += s[length]
        length -= 1
    return reversed_s

input_string = "Hello, world!"
reversed_string = reverse_string(input_string)
print(f"Original string: {input_string}")
print(f"Reversed string: {reversed_string}")
```

15.Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
def factorial_while_loop(number):
    """
    Calculate the factorial of a number using a while loop.

    Args:
        number: An integer for which the factorial is to be calculated.

    Returns:
        The factorial of the number, or 1 if the number is 0.
        Returns a message if the number is negative.
    """
    if number < 0:
        return "Factorial is not defined for negative numbers."
    elif number == 0:
        return 1
    else:
        factorial = 1
        while number > 0:
            factorial *= number
            number -= 1
        return factorial

if __name__ == "__main__":
    try:
        num = int(input("Enter a non-negative integer: "))
        result = factorial_while_loop(num)
        print(f"The factorial of {num} is {result}")
    except ValueError:
        print("Invalid input. Please enter an integer.")
```
